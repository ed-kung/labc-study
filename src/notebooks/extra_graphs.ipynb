```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 11
```

```python
data = np.array([
    ['Los Angeles',   59], 
    ['San Diego',     54.5], 
    ['Philadelphia',  54.5], 
    ['Indianapolis',  54], 
    ['NYC',           53],
    ['San Jose',      52],
    ['Houston',       51], 
    ['San Antonio',   50], 
    ['Jacksonville',  50], 
    ['Miami',         49.5],
    ['Fort Worth',    49],
    ['Chicago',       48], 
    ['Phoenix',       47.5],
    ['Austin',        47], 
    ['Denver',        47], 
    ['Charlotte',     47], 
    ['Dallas',        46.5],
    ['Washington DC', 46],
    ['Columbus',      44.5], 
    ['Seattle',       44]
])
data = pd.DataFrame(data=data, columns=['city', 'rentburden'])
data['rentburden'] = data['rentburden'].astype('float')

plt.figure(figsize=(6,2))
plt.rc('axes', labelsize=12)
plt.bar(data['city'], data['rentburden'])
plt.xticks(rotation=60, ha='right', rotation_mode='anchor')
plt.ylim(35, 60)
plt.grid(axis='y', alpha=0.2)
plt.gca().set_axisbelow(True)
plt.savefig('../../results/fig_rentburden.svg', bbox_inches='tight')
plt.show()
```

```python
data = np.array([
    [2015, 25686, 17687, 7999], 
    [2016, 28464, 21338, 7126],
    [2017, 33138, 24186, 8952],
    [2018, 31285, 22887, 8398], 
    [2019, 36165, 27221, 8944],
    [2020, 41290, 28852, 12438]
])
data = pd.DataFrame(data=data, columns=['year','total','unsheltered','sheltered'])

plt.figure(figsize=(6,3))
plt.rc('axes',labelsize=12)
plt.bar(data['year'], data['sheltered'])
plt.bar(data['year'], data['unsheltered'], bottom=data['sheltered'])
#plt.bar(data['year'], data['total'])
#plt.bar(data['year'], data['unsheltered'])
#plt.plot(data['year'], data['total'], 'o-')
#plt.plot(data['year'], data['unsheltered'], 'o-')
plt.legend(['Sheltered', 'Unsheltered'])
for i in range(len(data)):
    x = data.loc[i, 'year']
    y = data.loc[i, 'total'] + 600
    text = f"{data.loc[i,'total']:,.0f}"
    plt.text(x, y, text, ha='center', )
plt.grid(axis='y', alpha=0.2)
plt.ylim(0, 50000)
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_yticks()])
plt.gca().set_axisbelow(True)
plt.savefig('../../results/fig_homeless.svg', bbox_inches='tight')
plt.show()
```

```python
ACTUAL = 83865
RHNA = 456643

data = np.array([
    ['RHNA Allocation\n2021-2029', RHNA],
    ['Dwelling units\nadded 2010-2019', ACTUAL], 
])
data = pd.DataFrame(data=data, columns=['type', 'total'])
data['total'] = data['total'].astype('float')
data['baseline'] = ACTUAL
data['needed'] = data['total'] - ACTUAL

plt.figure(figsize=(5.5,2))
plt.rc('axes',labelsize=12)
plt.barh(data['type'], data['baseline'])
plt.barh(data['type'], data['needed'], left=data['baseline'], color='#1f97d4')
plt.text(ACTUAL + 5000, 1, f"{ACTUAL:,.0f}", va='center')
plt.text(RHNA + 5000, 0, f"{RHNA:,.0f}", va='center')
plt.text((RHNA-ACTUAL)/2 + ACTUAL, 0, f"Difference = {RHNA-ACTUAL:,.0f}", color='white', ha='center', va='center')
plt.xlim(0, 525000)
plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in plt.gca().get_xticks()])
plt.grid(axis='x',alpha=0.2)
plt.gca().set_axisbelow(True)
plt.savefig('../../results/fig_rhna.svg', bbox_inches='tight')
plt.show()
```
